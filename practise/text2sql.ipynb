```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import PydanticOutputParser
from pydantic import BaseModel
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model=ChatGroq(model="llama-3.1-8b-instant",temperature=0)
```

```python
# Define the input template:
# Define the output template:
# Send it to LLM:
# Get an Ouput
```

```python
class SQLTemplate(BaseModel):
    sql_query:str
```

```python
parser=PydanticOutputParser(pydantic_object=SQLTemplate)
```

```python
prompt = PromptTemplate(
    template="""
You are an expert SQL generator.

Your task:
- Generate ONLY a SQL QUERY STRING
- Wrap that SQL query inside a JSON object

Rules:
- Do NOT use SQL JSON functions (json_build_object, json_object, etc.)
- Do NOT explain anything
- Do NOT output SQL outside JSON
- Do NOT include markdown or code blocks

The JSON must match this schema exactly:
{format}

Request:
{text}
""",
    input_variables=["text"],
    partial_variables={
        "format": parser.get_format_instructions()
    }
)
```

```python
chain=prompt|model|parser
```

```python
chain.invoke({'text':'Give a general query to choose the all the rows and columns from a table'})
```
